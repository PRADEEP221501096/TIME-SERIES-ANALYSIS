```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "monthly-beer.csv"
data = pd.read_csv(file_path)

# Check the first few rows
print(data.head(25))

# Convert the date column (if exists) to a datetime type
data['Month'] = pd.to_datetime(data['Month'])  # Adjust column name if different

# Calculate Q1 (25th percentile) and Q3 (75th percentile) for the 'Monthly beer production' column
Q1 = data['Monthly beer production'].quantile(0.25)
Q3 = data['Monthly beer production'].quantile(0.75)

# Calculate IQR (Interquartile Range)
IQR = Q3 - Q1

# Determine the lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = data[(data['Monthly beer production'] < lower_bound) | (data['Monthly beer production'] > upper_bound)]

# Print outliers if any
if not outliers.empty:
    print("Outliers detected:")
    print(outliers)
else:
    print("No outliers detected.")

# Plot the data with outliers highlighted
plt.figure(figsize=(25, 4))
plt.plot(data['Month'], data['Monthly beer production'], label='Beer Production')
plt.scatter(outliers['Month'], outliers['Monthly beer production'], color='red', label='Outliers', zorder=5)
plt.title('Monthly Beer Production with Outliers')
plt.xlabel('Date')
plt.ylabel('Beer Production')
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "/content/monthly-beer.csv"
data = pd.read_csv(file_path)

# Check the first few rows
print(data.head(25))

# Convert the date column (if exists) to a datetime type
data['Month'] = pd.to_datetime(data['Month'])  # Adjust column name if different

# Calculate Q1 (25th percentile) and Q3 (75th percentile) for the 'Monthly beer production' column
Q1 = data['Monthly beer production'].quantile(0.25)
Q3 = data['Monthly beer production'].quantile(0.75)

# Calculate IQR (Interquartile Range)
IQR = Q3 - Q1

# Determine the lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = data[(data['Monthly beer production'] < lower_bound) | (data['Monthly beer production'] > upper_bound)]

# Print outliers if any
if not outliers.empty:
    print("Outliers detected:")
    print(outliers)
else:
    print("No outliers detected.")

# Remove the outliers from the dataset
data_no_outliers = data[(data['Monthly beer production'] >= lower_bound) & (data['Monthly beer production'] <= upper_bound)]

# Plot before removing outliers (Bar plot)
plt.figure(figsize=(12, 6))
plt.bar(data['Month'], data['Monthly beer production'], color='blue', label='Before Removing Outliers')
plt.title('Monthly Beer Production - Before Removing Outliers')
plt.xlabel('Date')
plt.ylabel('Beer Production')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.show()

# Plot after removing outliers (Bar plot)
plt.figure(figsize=(12, 6))
plt.bar(data_no_outliers['Month'], data_no_outliers['Monthly beer production'], color='green', label='After Removing Outliers')
plt.title('Monthly Beer Production - After Removing Outliers')
plt.xlabel('Date')
plt.ylabel('Beer Production')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.show()
```
