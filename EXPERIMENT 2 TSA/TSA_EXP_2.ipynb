```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "monthly-beer.csv"
df = pd.read_csv(file_path)

# Convert 'Month' to datetime for proper plotting
df["Month"] = pd.to_datetime(df["Month"])

# Set plot style
sns.set_theme(style="darkgrid")

# Create figure
plt.figure(figsize=(10, 6))
plt.title("Bar Plot: Yearly Beer Production")

# Bar plot
sns.barplot(x=df["Month"].dt.year, y=df["Monthly beer production"], ci=None)
plt.xlabel("Year")
plt.ylabel("Beer Production")
plt.xticks(rotation=45)

plt.show()
```
